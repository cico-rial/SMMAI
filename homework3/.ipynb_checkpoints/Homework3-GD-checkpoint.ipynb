### Backtracking and GD definition
`Note:` the backtracking and GD function have been modified by Leonardo Chiarioniâ„¢ introducing a new parameter: **multivariate**

```python
def backtracking(f, grad_f, x, multivariate):
    """
    This function is a simple implementation of the backtracking algorithm for
    the GD (Gradient Descent) method.
    
    f: function. The function that we want to optimize.
    grad_f: function. The gradient of f(x).
    x: ndarray. The actual iterate x_k.
    multivariate: boolean. Indicate whether the function f 
    is multivariate or not.
    """

    if multivariate:
        norm = np.linalg.norm
    else:
        norm = np.abs
        
    alpha = 1
    c = 0.8
    tau = 0.25
    
    while f(x - alpha * grad_f(x)) > f(x) - c * alpha * norm(grad_f(x)) ** 2:
        alpha = tau * alpha
        
        if alpha < 1e-3:
            break
    return alpha



def GD(f, grad_f, x0, kmax, tolf, tolx, x_true, back_tracking=False, alpha=1, multivariate=True, plot=False):
    """
    This function implements the Gradient Descent algorithm.
    """
    if multivariate:
        norm = np.linalg.norm
    else:
        norm = np.abs
    
    # Initialize x_k
    x_k = x0
    
    # counter
    k = 0
    
    # Initialize the outputs
    x = [x0]
    f_val = [f(x0)]
    grads = [grad_f(x0)]
    err_grad = [norm(grad_f(x0))]
    err_sol = [norm(x_true - x0)]
    
    if plot:
        x_range = np.linspace(-10, 10, 400)
        y_range = np.linspace(-10, 10, 400)
        X, Y = np.meshgrid(x_range, y_range)
        Z = f((X, Y))

        plt.figure(figsize=(5, 5))
        plt.contour(X, Y, Z, levels=np.logspace(-2, 3, 20), cmap='viridis')
        plt.scatter(x0[0], x0[1], color='red', marker='o', label='Initial Point')
    
    # Loop
    condition = True
    while condition:
        # Update alpha
        if back_tracking:
            alpha = backtracking(f, grad_f, x_k, multivariate)
        # Update x
        x_k = x_k - alpha*grad_f(x_k)
        
        # Update outputs
        x.append(x_k)
        f_val.append(f(x_k))
        grads.append(grad_f(x_k))
        err_grad.append(norm(grad_f(x_k)))
        err_sol.append(norm(x_true - x_k))
        
        if plot:
            plt.scatter(x_k[0], x_k[1], color='blue', marker='o', label=f'Iteration {k}')
               
        # check criteria
        condition1 = norm(grad_f(x_k)) > (tolf * norm(grad_f(x0)))
        condition2 = norm(x_k) > tolx
        condition3 = k < kmax
        condition = condition1 and condition2 and condition3
       
        # update k
        k += 1
        
    if plot:
        plt.title('Gradient Descent Iterations')
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.show()

    return x_k, k, np.array(x), np.array(f_val), np.array(grads), np.array(err_grad), np.array(err_sol)

def plot_errors(x, k, err_grad, err_sol, text, figsize=(10,3)):
    """
    This function enables you to simply plot the error 
    of the gradient descent with respect to the iteration k
    """
    print(text)
    print(f'x = {x[-1]}')
    print(f"k: {k}")
    print(f'gradient error: {err_grad[k]}')
    print(f'solution error: {err_sol[k]}')
    
    plt.figure(figsize=figsize)
    plt.subplot(1,2,1)
    plt.plot(np.arange(1,k+2,1), err_grad, 'r-')
    plt.xlabel('k-th iteration')
    plt.title('Gradient error')
    plt.grid()
    
    plt.subplot(1,2,2)
    plt.plot(np.arange(1,k+2,1), err_sol, 'r-')
    plt.xlabel('k-th iteration')
    plt.title('Solution Error')
    plt.grid()
    
    plt.show()
```

### Function 1:


```python
def f(x):
    x1 = x[0]
    x2 = x[1]
    return (x1 - 3)**2 + (x2 - 1)**2

def grad_f(x):
    x1 = x[0]
    x2 = x[1]
    return np.array([2*(x1 - 3), 2*(x2 - 1)])
```

Backtracking = **True**, convergence in 87 iterations

```python
x0 = np.array([0,0])
kmax = 1_00
tolf = tolx = 1e-5
x_true = np.array([3,1])

x_sol, k, x, f_val, grads, err_grad, err_sol = GD(f, grad_f, x0, kmax, tolf, tolx, x_true, back_tracking=True, plot=True)
plot_errors(x, k, err_grad, err_sol, "back_tracking = True")
```

Backtracking = **False**, high learning rate: **Divergence!**

```python
x_sol, k, x, f_val, grads, err_grad, err_sol = GD(f, grad_f, x0, kmax, tolf, tolx, x_true, alpha=1.1)
plot_errors(x, k, err_grad, err_sol, 'alpha = 1.1')
```

Backtracking = False, correct learning rate: just **1 iteration!**

```python
x_sol, k, x, f_val, grads, err_grad, err_sol = GD(f, grad_f, x0, kmax, tolf, tolx, x_true, alpha=0.5)
plot_errors(x, k, err_grad, err_sol, 'alpha = 0.5')
```

### Function 3:

Backtracking is superslow with this function. Experimentally, a step-size greater than about 0.22 leads to divergence. A step size of 0.2 enables us to reach an optimal solution in 200 iterations.

```python
def f(x):
    n = x.shape[0]
    v = np.linspace(0,1,n)
    x_true = np.ones((n,))
    A = np.vander(v, increasing=True)
    b = A @ x_true
    return 0.5*np.linalg.norm(A@x_true-b)**2

def grad_f(x):
    n = x.shape[0]
    v = np.linspace(0,1,n)
    x_true = np.ones((n,))
    A = np.vander(v, increasing=True)
    b = A @ x_true
    return A.T@(A@x-b)
```

```python
n = 5
x0 = np.zeros(n)
kmax = 1_000
tolf = tolx = 1e-5
x_true = np.ones((n,))


x_sol, k, x, f_val, grads, err_grad, err_sol = GD(f, grad_f, x0, kmax, tolf, tolx, x_true, back_tracking=True, multivariate=True)
plot_errors(x, k, err_grad, err_sol, "n = 5\nback_tracking = True")
```

```python
x_sol, k, x, f_val, grads, err_grad, err_sol = GD(f, grad_f, x0, kmax, tolf, tolx, x_true, alpha=0.2, multivariate=True)
plot_errors(x_sol, k, err_grad, err_sol, f"n = {n}\nalpha = 0.2")
```

### Function 4:

Backtracking doesn't allow convergence here (high error at the end). A step-size greater than about 0.2 leads to divergence

```python
l = 0.5
def f(x):
    n = x.shape[0]
    v = np.linspace(0,1,n)
    x_true = np.ones((n,))
    A = np.vander(v, increasing=True)
    b = A @ x_true
    return 0.5*np.linalg.norm(A@x_true-b)**2 + 0.5*l*np.linalg.norm(x)**2

def grad_f(x):
    n = x.shape[0]
    v = np.linspace(0,1,n)
    x_true = np.ones((n,))
    A = np.vander(v, increasing=True)
    b = A @ x_true
    return A.T@(A@x-b) + l*x
```

```python
n = 5
x0 = np.zeros(n)
kmax = 1_00
tolf = tolx = 1e-5
x_true = np.ones((n,))

x_sol, k, x, f_val, grads, err_grad, err_sol = GD(f, grad_f, x0, kmax, tolf, tolx, x_true, back_tracking=True, multivariate=True)
plot_errors(x, k, err_grad, err_sol, f'l = {l}\nn = {n}\nback_tracking = True')
```

```python
x_sol, k, x, f_val, grads, err_grad, err_sol = GD(f, grad_f, x0, kmax, tolf, tolx, x_true, alpha=0.2, multivariate=True)
plot_errors(x, k, err_grad, err_sol, f'l = {l}\nn = {n}\nalpha = 0.2')
```

### Function 5:

Backtracking is useful to avoid non optimal solution or even divergence, unless a very tiny step-size is used. If the starting point is towards the negative side, the other (local) minimum will be found.

```python
def f(x):
    return x**4 + x**3 -2*x**2 -2*x

def grad_f(x):
    return 4*x**3 + 3*x**2 + -4*x -2
```

```python
x0 = 0
kmax = 1_00
tolf = tolx = 1e-5
x_true = 0.9222247929324 # computed using geogebra

x_sol, k, x, f_val, grads, err_grad, err_sol = GD(f, grad_f, x0, kmax, tolf, tolx, x_true, back_tracking=True, multivariate=False)
plot_errors(x, k, err_grad, err_sol, 'back_tracking=True')

a,b = -2, 2
x_val = np.linspace(a,b,200)
plt.figure(figsize=(10,4))
plt.scatter(x, f(x), c='black', marker='x', s=20, label='minimum')
plt.plot(x_val, f(x_val), 'r-', label='f(x)')
plt.grid()
plt.legend()
plt.show()
```
