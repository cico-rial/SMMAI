## Visualizing dyads

Consider an image from skimage.data. For simplicity, say that $X ∈ R^{m \times n}$ is the matrix representing
that image. You are asked to visualize the dyad of the SVD Decomposition of X and the result of compressing
the image via SVD. In particular:

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage;
```

• Load the image into memory and compute its SVD;

```python
x = skimage.data.camera()
m, n = x.shape

plt.imshow(x, cmap="gray")
plt.axis("off")
plt.show()

U, s, VT = np.linalg.svd(x)
S = np.zeros((m,n)) 
S[:n, :n] = np.diag(s)
```

• Visualize some of the dyad $\sigma_i u_i v_i^T$
of this decomposition. What do you notice?

```python
plt.figure(figsize=(10, 5))

for j,i in enumerate([1,2,5,10,20,50,100,200,350,500]):   
    ui = U[:, i:i+1] 
    vit = VT[i:i+1, :] 
    dyad_i = s[i]*(ui@vit)
    
    plt.subplot(2,5,j+1)
    plt.title(f'dyad {i}')
    plt.imshow(dyad_i, cmap='gray') # because it is in grey scale
    plt.axis("off")
    
plt.show()
```

The dyads shows that relevant patterns of the image are captured by the first few singular vector, which are weighted by the correspondent singular values!

• Plot the singular values of X. Do you note something?

```python
#Plot singular values of X
plt.figure(figsize=(10, 3))
plt.plot(s)
plt.title('Singular values')
plt.grid()
plt.show()
```

As we know from the theory, singular values are arranged in descending order!

• Visualize the k-rank approximation of X for different values of k. What do you observe?

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
j = 0
for i,k in enumerate([1,20,40,60,80,100]):
    if i % 3 == 0 and i != 0:
        j += 1
    axs[j,i%3].set_title(f'Rank {k} approximation') 

    X_k = U[:, :k] @ S[:k, :k] @ VT[:k, :]

    axs[j,i%3].imshow(X_k, cmap='gray') # because it is in grey scale
    axs[j,i%3].axis("off")
    
    caption = f"Rank of X_k: {np.linalg.matrix_rank(X_k)}\n" + f"||X - X_k||_2 = {round(np.linalg.norm(x - X_k, 2),2)}"
    
    axs[j,i%3].text(0, 560, caption, fontsize=8, color='black')

plt.show()
```

Compute and plot the approximation error $||X − X_k||_F$ for increasing values of k, where $X_k$ is the
k-rank approximation of k.
F stays for Frobenius Norm

```python
k_numbers = np.arange(1,513,1)
compression_factors = np.ones(k_numbers.shape)
approx_errors = np.zeros(k_numbers.shape)
for k in k_numbers:
    X_k = U[:, :k] @ S[:k, :k] @ VT[:k, :]
    approx_errors[k-1] = np.linalg.norm(x - X_k, 2)
    compression_factors[k-1] = k*(m+n+1)/(m*n)
plt.plot(k_numbers, approx_errors)
plt.grid()
plt.title("Approximation errors")
plt.xlabel("k")
plt.ylabel("||X - X_k||_F")
plt.show()
```

• Plot the compression factor $ck = 1 - \frac{k(m+n+1)}{mn}$
for increasing k.

```python
plt.plot(k_numbers, 1 - compression_factors)
plt.grid()
plt.title("compression factor")
plt.xlabel("k")
plt.ylabel("k*(1+m+n)/m*n")
plt.show()
```

• What is the approximation error when
the compressed image requires the same amount of informations of those of the uncompressed image 
(i.e. ck = 0)?

```python
i = np.where((1 - compression_factors) <= 0)[0][0]

X_i = U[:, :i] @ S[:i, :i] @ VT[:i, :]
plt.title(f"Rank = {i}, approx_error = {approx_errors[i]}, c_k = {1 - compression_factors[i]}", )
plt.imshow(X_i, cmap="gray")
plt.axis("off")
plt.show()
```
